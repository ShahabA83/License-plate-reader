```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import os
```

```python
path = 'C:/Users/Shahab/Downloads/Jupyter/Car-license-Plate/My-Code/Project/Cutting-pieces-of-letters-and-numbers-on-the-license-plate/data'
img1 = cv.imread(os.path.join(path, "license-plate-digital1.png"), 0)
img2 = cv.imread(os.path.join(path, "license-plate-digital2.png"), 0)
img3 = cv.imread(os.path.join(path, "license-plate-digital3.png"), 0)
img4 = cv.imread(os.path.join(path, "license-plate-digital4.png"), 0)
```

```python
fig , ax = plt.subplots(4, figsize=(12,8))
ax[0].imshow(img1, cmap='gray')
ax[1].imshow(img2, cmap='gray')
ax[2].imshow(img3, cmap='gray')
ax[3].imshow(img4, cmap='gray')
```

```python
lst_digital = [img1, img2, img3, img4]
chopfactors = [(60,125), (125,190), (180,300), (280,350), (340,415), (400,465), (460,535), (515,600)]
```

```python
for i in lst_digital:
    h, w = i.shape
    fig , ax = plt.subplots(1,8, figsize=(12,8))
    plate_letters = []
    c = 0
    for factor in chopfactors:
        w1 = int((factor[0]/600)*w)
        w2 = int((factor[1]/600)*w)
        letterpatch = i[:,w1:w2]
        letterpatch_resized  = cv.resize(letterpatch, (28,28), interpolation = cv.INTER_LINEAR)
        ax[c].imshow(letterpatch_resized, cmap='gray')
        c += 1
        plate_letters.append(letterpatch_resized)
```
